```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import re
```

```python
def isNaN(text):
    return text!=text
```

```python
# 크롤링한 데이터 불러오기
data = pd.read_csv('Data/크림_조던_거래량_데이터1700_to_2000.csv')
data.head()
```

```python
data.info()
```

# 1. 제품 정보

```python
# 제품 정보를 담은 데이터만 남기고 중복 제거 
data_info = data[data['제품한글명'].notnull()]
data_info = data_info.drop_duplicates(subset=['제품한글명'])
data_info = data_info.reset_index(drop=True)
data_info.head()
```

```python
# 필요한 칼럼 추가 
data_info = data_info[['index','제품한글명','제품정보리스트']]
data_info.insert(3,'모델번호',None)
data_info.insert(4,'출시일',None)
data_info.insert(5,'컬러',None)
data_info.insert(6,'발매가',None)
```

```python
data_info.head()
```

```python
# split해서 칼럼 저장 
for idx,info in enumerate(tqdm(data_info['제품정보리스트'])):
    if isNaN(info):
        pass
    else:
        info_list = info.split('\n')

        for i in range(len(info_list)):
            if info_list[i]=='모델번호':
                data_info.loc[idx,'모델번호'] = info_list[i+1]
            if info_list[i]=='출시일':
                data_info.loc[idx,'출시일'] = info_list[i+1]
            if info_list[i]=='컬러':
                data_info.loc[idx,'컬러']= info_list[i+1]
            if info_list[i]=='발매가':
                data_info.loc[idx,'발매가'] = info_list[i+1]
```

```python
data_info.head()
```

```python
# 데이터 저장 
data_info = data_info.drop('제품정보리스트',axis=1)
data_info.to_csv('Data/크림_조던_제품정보_1700_2000.csv',index=False,encoding='utf-8-sig')
```

# 2. 거래량

```python
# 거래 데이터만 남기기
data_trans = data[data['거래리스트'].notnull()]
data_trans = data_trans.drop_duplicates()
data_trans = data_trans.reset_index(drop=True)
data_trans = data_trans.drop(['제품한글명','제품정보리스트'],axis=1)
data_trans.head()
```

```python
# 필요 칼럼 추가
data_trans.insert(2,'사이즈',None)
data_trans.insert(3,'거래가',None)
data_trans.insert(4,'거래일',None)
data_trans.head()
```

```python
# split해서 칼럼 저장 
for idx,info in enumerate(tqdm(data_trans['거래리스트'])):
    info_list = info.split('\n')
    
    data_trans.loc[idx,'사이즈'] = info_list[0]
    data_trans.loc[idx,'거래가'] = info_list[1]
    if info_list[2]=='빠른배송':
        data_trans.loc[idx,'거래일'] = info_list[3]
    else:
        data_trans.loc[idx,'거래일'] = info_list[2]
    
```

```python
data_trans = data_trans.drop('거래리스트',axis=1)
data_trans.head()
```

# 3. merge

```python
# 제품 정보랑 제품 거래량 merge
data_trans_final = pd.merge(data_trans,data_info,on='index',how='left')
data_trans_final.head()
```

```python
# 데이터 저장 
data_trans_final.to_csv('Data/크림_조던_제품거래량_1700_2000.csv',index=False,encoding='utf-8-sig')
```

# 4. 발매가 형식 통일

```python
data = pd.read_csv('Data/크림_조던_제품거래량_2000_3000.csv')
data.head()
```

```python
# 발매가 null 값 제거
data = data.dropna(subset=['발매가'])
data = data.reset_index(drop=True)
data.info()
```

```python
for idx,money in enumerate(data['발매가']):
    if '(' in money:
        data.loc[idx,'발매가'] = re.search(r'[0-9]+,[0-9]+원',money).group()
    else:
        pass
```

```python
data.head(30)
```

```python
data.to_csv('Data/크림_조던_제품거래량_발매가가공_2000_3000.csv',index=False,encoding='utf-8-sig')
```
